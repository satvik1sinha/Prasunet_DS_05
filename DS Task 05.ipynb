```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a synthetic dataset
np.random.seed(0)
sample_size = 1000
data = {
    'date': pd.date_range(start='2023-01-01', periods=sample_size, freq='H'),
    'road_condition': np.random.choice(['Dry', 'Wet', 'Icy', 'Snowy'], sample_size),
    'weather': np.random.choice(['Clear', 'Rain', 'Fog', 'Snow'], sample_size),
    'time_of_day': np.random.choice(['Morning', 'Afternoon', 'Evening', 'Night'], sample_size),
    'accident_severity': np.random.choice(['Minor', 'Major', 'Fatal'], sample_size),
    'location': np.random.choice(['Location A', 'Location B', 'Location C', 'Location D'], sample_size)
}
df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```

```python
# Data Cleaning
# Convert 'date' to datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract hour and day of week from date
df['hour'] = df['date'].dt.hour
df['day_of_week'] = df['date'].dt.day_name()

# Display basic information about the dataset
df.info()
```

```python
# Display summary statistics of the dataset
print(df.describe(include='all'))
```

# Visualizations

```python
# Distribution of accidents by road condition
plt.figure(figsize=(10, 6))
sns.countplot(x='road_condition', data=df, palette='viridis')
plt.title('Accidents by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of accidents by weather condition
plt.figure(figsize=(10, 6))
sns.countplot(x='weather', data=df, palette='viridis')
plt.title('Accidents by Weather Condition')
plt.xlabel('Weather')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of accidents by time of day
plt.figure(figsize=(10, 6))
sns.countplot(x='time_of_day', data=df, palette='viridis')
plt.title('Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of accidents by day of week
plt.figure(figsize=(10, 6))
sns.countplot(x='day_of_week', data=df, palette='viridis', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Accidents by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.show()
```

```python
# Accident severity by road condition
plt.figure(figsize=(10, 6))
sns.countplot(x='road_condition', hue='accident_severity', data=df, palette='viridis')
plt.title('Accident Severity by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Count')
plt.show()
```

```python
# Accident severity by weather condition
plt.figure(figsize=(10, 6))
sns.countplot(x='weather', hue='accident_severity', data=df, palette='viridis')
plt.title('Accident Severity by Weather Condition')
plt.xlabel('Weather')
plt.ylabel('Count')
plt.show()
```

```python
# Accident hotspots
plt.figure(figsize=(10, 6))
sns.countplot(x='location', data=df, palette='viridis')
plt.title('Accident Hotspots')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()
```
